# **누락된 값 처리하기**

데이터프레임에 누락된 값이 있다는 것을 대략 확인하였으니 정확한 데이터 분석을 위해 누락된 값을 다루는 방법을 자세히 알아본다. 각 열의 누락된 값이 정확하게 몇 개가 있는지 확인하고, 누락된 값을 표시하는 방법을 알아본다. 그리고 그 값을 다른 값으로 바꾸어 본다.

info()메서드로 출력한 값으로 누락된 값이 있는 행의 개수를 헤아릴 수 있지만, 번거롭다. 대신 NaN을 직접 카운트할 수 있는 isna()메서드를 사용하면 훨씬 편리하다. isna()메서드는 각 행이 비어 있는지를 나타내는 불리언 배열을 반환한다. 그리고 sum()메서드를 이어서 호출하면 불리언 배열의 True 개수로 비어 있는 행 개수를 얻을 수 있다.

```python
ns_book4.isna().sum()
```

열 데이터 타입은 int64로 64비트 정수를 저장할 수 있다. 실행 결과를 보니 '도서권수' 열은 누락된 값이 없다. isna()메서드와 반대로 누락되지 않은 값을 확인할 때는 notna()메서드를 사용한다.

###누락된 값으로 표시하기: None과 np.nan

이번에는 임의로 누락된 값을 만든 후 NaN으로 표시하는 방법을 알아본다. 판다스 데이터프레임에서는 정수를 저장하는 열에 파이썬의 None을 입력하면 누락된 값으로 인식한다.

ns_book4 데이터프레임의 '도서권수' 열에 있는 첫 번째 행 값을 None로 바꾼 후, '도서권수' 열에 isna()메서드를 적용해 누락된 값을 세어 보면, 다음과 같다.

```python
ns_book4.loc[0, '도서권수'] = None
ns_book4['도서권수'].isna().sum()
```

head()메서드로 처음 두 개의 행을 출력하여 첫 번째 행의 '도서권수' 값이 어떻게 바뀌었나 확인한다.

```python
ns_book4.head(2)
```

NaN이 출력된다. 그런데 두 번째 행의 '도서권수' 값이 1이었는데 1.0으로 바뀌었다. 이는 판다스가 NaN을 특별한 실수 값으로 저장하기 때문이다. 그래서 원래 데이터타입이 inta64였던 '도서권수' 열이 NaN을 표시하기 위해 float64로 자동으로 변환된 것이다.

'도서권수' 열의 첫 번째 행을 원래대로 1로 바꾸어 본다. 현재 데이터 타입이 실수형이므로 다시 정수형으로 바꿔야 한다. 이때 '도서권수' 열과 함께 '대출건수'열도 데이터 타입을 int64로 바꾸어 보겠다. 데이터 타입을 지정할 때는 astype()메서드를 사용한다. 매개변수를 {열 이름: 데이터 타입} 형식의 딕셔너리로 전달한다.

```python
ns_book4.loc[0, '도서권수'] = 1
ns_book4 = ns_book4.astype({'도서권수': 'int64', '대출건수': 'int64'})
ns_book4.head(2)
```

정수형이 아니라 문자열을 저장할 수 있는 열에 None을 입력하면 어떨까? 데이터 타입이 object인 '부가기호' 열의 첫 번째 행에 None을 입력해 보겠다.

```python
ns_book4.loc[0, '부가기호'] = None
ns_book4.head(2)
```

정수형 열에 입력했을 때처럼 NaN으로 표시되지 않고 문자열 그대로 None으로 표시된다. 그럼 None 대신에 NaN을 입력해야 할까?

사실 판다스는 NaN이라는 값을 따로 가지고 있지 않다. 대신 넘파이 패키지에 있는 np.nan을 사용해야 한다. 따라서 첫 번째 행의 '부가기호' 열의 값을 None에서 NaN으로 표시하려면 다음과 같이 np.nan을 사용한다.

```python
import numpy as np
ns_book4.loc[0, '부가기호'] = np.nan
ns_book4.head(2)
```

##누락된 값 바꾸기(1): loc, fillna()메서드

'세트 ISBN' 열은 대부분의 값이 비어 있다. 이 누락된 값을 NaN이 아니라 빈 문자열('')로 바꾸어 보겠다.

loc메서드를 사용하면 누락된 값을 원하는 값으로 바꿀 수 있다. 그러려면 누락된 값을 가리키는 불리언 배열을 만들어야 하는데, 누락된 값을 확인하는 isna()메서드로 간단하게 만들 수 있다.

isna()메서드로 '세트 ISBN' 열의 NaN을 가리키는 set_isbn_na_rows 불리언 배열을 만들고, loc메서드에 전달한다.

```python
set_isbn_na_rows = ns_book4['세트 ISBN'].isna()
ns_book4.loc[set_isbn_na_rows, '세트 ISBN'] = ''
ns_book4['세트 ISBN'].isna().sum()
```

'세트 ISBN' 열의 NaN을 모두 빈 문자열로 바꾸었기 때문에 누락된 행의 개수는 0이 된다.

이처럼 loc메서드로 NaN을 다른 값으로 바꿀 수 있지만, 조금 더 편리한 방법은 fillna()메서드이다. fillna()메서드에 원하는 값을 전달하면 NaN을 대체할 수 있다.

ns_book4에 있는 모든 NaN을 '없음' 문자열로 바꾸어 보겠다. fillna()메서드는 기본적으로 새로운 데이터프레임을 반환하기 때문에, fillna()메서드에 이어 isna()메서드를 연결하면 NaN의 개수를 셀 수 있다.

```python
ns_book4.fillna('없음').isna().sum()
```

NaN이 모두 '없음' 문자열로 바뀌면서 NaN 개수를 세면 0이 출력된다. 물론 다음 코드처럼 특정 열만 선택해서 NaN으로 바꿀 수도 있다.

```python
ns_book4['부가기호'].fillna('없음').isna().sum()
```

하지만 특정 열을 선택한 후 fillna()메서드를 적용하면 앞의 실행 결과처럼 열 이름 없이 개수만 있는 판다스 시리즈 객체로 반환한다.

'부가기호' 열의 NaN을 바꾸면서 전체 데이터프레임을 반환하려면 다음처럼 열 이름과 바꾸려는 값으로 이루어진 딕셔너리를 전달하면 된다.

```python
ns_book4.fillna({'부가기호': '없음'}).isna().sum()
```

##누락된 값 바꾸기(2): replace()메서드

replace()메서드는 NaN은 물론 어떤 값도 바꿀 수 있는 편리한 메서드이다. replace()메서드는 사용법이 다양하다는 특징이 있다.

###첫째, 바꾸려는 값이 하나일 때

바꾸려는 값이 하나라면 replace()메서드의 첫 번째 매개변수에는 원래 값을, 두 번째 매개변수에는 새로운 값을 전달한다. replace(원래 값, 새로운 값)

원래 값에 np.nan을 전달하여 NaN을 '없음'으로 바꿔 보겠다.

```python
ns_book4.replace(np.nan, '없음').isna().sum()
```

###둘째, 바꾸려는 값이 여러 개일 때

바꾸려는 값이 여러 개일 때는 리스트 형식으로 전달한다. replace([원래 값1, 원래 값2], [새로운 값1, 새로운 값2])

NaN을 '없음'으로 바꾸고 '2021' 문자열을 '21'로 바꾸면 다음과 같다.

```python
ns_book4.replace([np.nan, '2021'], ['없음', '21']).head(2)
```

또는 리스트 대신 ({원래 값1:새로운 값1, 원래 값2: 새로운 값2})처럼 딕셔너리 형식으로 전달할 수도 있다.

```python
ns_book4.replace({np.nan: '없음', '2021': '21'}).head(2)
```

###셋째, 열마다 다른 값으로 바꿀 때

열 이름과 바꾸려는 값을 딕셔너리 형식으로 전달하여 열마다 다른 값을 바꿀 수 있다. replace({열 이름: 원래 값}, 새로운 값)

'부가기호' 열의 NaN을 '없음'으로 바꾸면 다음과 같다.

```python
ns_book4.replace({'부가기호': np.nan}, '없음').head(2)
```

또는 열 이름과 변경 전후의 값을 ({열 이름: {원래 값1: 새로운 값1}})과 같이 중첩된 딕셔너리로 전달할 수 있다. '부가기호' 열의 NaN을 '없음'으로 바꾸고 '발행년도' 열의 '2021'을 '21'로 바꾸면 다음과 같다.

```python
ns_book4.replace({'부가기호': {np.nan: '없음'},
                           '발행년도': {'2021': '21'}}).head(2)
```

지금까지 fillna()메서드와 replace()메서드를 사용하여 누락된 값을 바꾸는 다양한 방법을 알아보았다. 마지막으로 replace() 메서드와 정규 표현식을 사용하는 방법을 알아본다.

#**정규 표현식**

정규 표현식 또는 줄여서 정규식은 문자열 패턴을 찾아서 대체하기 위한 규칙의 모임이다. replace()메서드로 인덱스 100번과 101번 행의 네 자리 연도('2021')를 두 자리('21')로 바꾸면 다음과 같다.

이때 연도가 다르면 적용되지 않아, 2018은 그대로인 것을 볼 수 있다.

```python
ns_book4.replace({'발행년도': {'2021': '21'}})[100:102]
```

물론 replace()메서드에 '2018'을 '18'로 바꾸도록 추가할 수 있다. 하지만 도서 발행 연도는 범위가 넓기 때문에 이런 식으로 일일이 모두 기재하여 바꾸는 행위는 번거롭다. 이럴 때 정규 표현식을 사용하면 훨씬 간편하게 작업을 수행할 수 있다.

```python
ns_book4.replace({'발행년도': {r'\d\d(\d\d)': r'\1'}}, regex=True)[100:102]
```

이때 정규 표현식 앞에 붙은 r문자는 원시문자열(raw string)을 만든다.

```python
ns_book4.replace({'발행년도': {r'\d{2}(\d{2})': r'\1'}}, regex=True)[100:102]
```

```python
ns_book4.replace({'저자': {r'(.*)\s\(지은이\)(.*)\s\(옮긴이\)': r'\1\2'},
                  '발행년도': {r'\d{2}(\d{2})': r'\1'}}, regex=True)[100:102]
```

#**잘못된 값 바꾸기**

먼저 데이터 타입을 지정해 보면, astype()메서드로 '발행년도' 열을 int64로 바꾸어 보니 오류가 발생한다.

```python
# 아래 코드는 오류가 발생함
ns_book4.astype({'발행년도': 'int32'})
```

'1988'이라는 연도를 변환할 수 없어 오류가 발생한 것으로 보인다. 숫자가 아닌 다른 문자가 들어 있는 것 같다. 정규 표현식으로 다른 문자가 있는 연도를 찾아본다.

판다스 시리즈 객체는 str 속성 아래 다양한 문자열 처리 함수를 제공한다. 그 중 contain()메서드는 시리즈나 인덱스에서 문자열 패턴을 포함하고 있는지 검사한다. '발행년도' 열에 '1988'이 포함된 행의 개수를 세어 보자.

```python
ns_book4['발행년도'].str.contains('1988').sum()
```

```python
invalid_number = ns_book4['발행년도'].str.contains(r'\D', na=True)
print(invalid_number.sum())
ns_book4[invalid_number].head()
```

```python
ns_book5 = ns_book4.replace({'발행년도': r'.*(\d{4}).*'}, r'\1', regex=True)
ns_book5[invalid_number].head()
```

```python
unknown_year = ns_book5['발행년도'].str.contains(r'\D', na=True)
print(unknown_year.sum())
ns_book5[unknown_year].head()
```

1777개에서 67개로 줄였다. 아직도 변환되지 않은 값은 NaN이거나 네 자리 숫자가 아닌 값이다. 이런 값은 어떻게 변환할지 알 수 없기 때문에 임의로 -1로 바꾼 다음 astype()메서드로 '발행년도' 열의 데이터 타입을 정수형인 int64로 변환한다.

```python
ns_book5.loc[unknown_year, '발행년도'] = '-1'
ns_book5 = ns_book5.astype({'발행년도': 'int64'})
```

연도 중에 이상하게 아주 큰 값이나 작은 값이 들어 있는 경우가 있다. 먼저 연도가 4000년이 넘는 경우를 확인해보자.

gt()메서드는 전달된 값보다 큰 값을 찾는다. 연도가 4000년이 넘는 행의 전체 개수를 확인하기 위해 sum()함수를 함께 사용한다.

```python
ns_book5['발행년도'].gt(4000).sum()
```

4000년이 넘는 경우가 131이나 된다. 아마 단군기원을 사용한 것으로 보인다.

```python
(ns_book5['발행년도'] > 4000).sum()
```

```python
dangun_yy_rows = ns_book5['발행년도'].gt(4000)
ns_book5.loc[dangun_yy_rows, '발행년도'] = ns_book5.loc[dangun_yy_rows, '발행년도'] - 2333
dangun_year = ns_book5['발행년도'].gt(4000)
print(dangun_year.sum())
ns_book5[dangun_year].head()
```

연도가 이상하게 높은 도서들은 모두 -1로 표시한다.

```python
ns_book5.loc[dangun_year, '발행년도'] = -1
```

마지막으로 연도가 작은 값도 확인하여야 한다. 이번엔 0보다 크고 1900보다 작은 연도의 도서를 찾아보겠다.

```python
old_books = ns_book5['발행년도'].gt(0) & ns_book5['발행년도'].lt(1900)
ns_book5[old_books]
```

여전히 잘못된 값이 있다. 이들도 연도를 -1로 설정하고 전체 행 개수를 확인해 보겠다.

```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_book_title(isbn):
  # Yes24 도서 검색 페이지 URL
  url = 'http://www.yes24.com/Product/Search?domain=BOOK&query={}'
  # URL에 ISBN을 넣어 HTML 가져오기
  r = requests.get(url.format(isbn))
  soup = BeautifulSoup(r.text, 'html.parser') # HTML 파싱
  # 클래스 이름이 'gd_name'인 <a> 태그의 텍스트 가져오기
  title = soup.find('a', attrs = {'class':'gd_name'}).get_text()
  return title
```

```python
import re

def get_book_info(row):
  title = row['도서명']
  author = row['저자']
  pub = row['출판사']
  year = row['발행년도']
  # Yes24 도서 검색 페이지 URL
  url = 'http://www.yes24.com/Product/Search?domain=BOOK&query={}'
  # URL에 ISBN을 넣어 HTML 가져오기
  r = requests.get(url.format(row['ISBN']))
  soup = BeautifulSoup(r.text, 'html.parser') # HTML 파싱
  try:
    if pd.isna(title):
      # 클래스 이름이 'gd_name'인 <a> 태그의 텍스트 가져오기
      title = soup.find('a', attrs = {'class': 'gd_name'}).get_text()
  except AttributeError:
    pass
  try:
    if pd.isna(author):
      # 클래스 이름이 'info_auth'인 <span> 태그 아래 <a> 태그의 텍스트 가져오기
      authors = soup.find('span', attrs = {'class': 'info_auth'}).find_all('a')
      author_list = [auth.get_text() for auth in authors]
      author = ','.join(author_list)
  except AttributeError:
    pass
  try:
    if pd.isna(pub):
      # 클래스 이름이 'info_pub'인 <span> 태그 아래 <a> 태그의 텍스트 가져오기
      pub = soup.find('span', attrs = {'class': 'info_pub'}).find('a').get_text()
  except AttributeError:
    pass
  try:
    if year == -1:
      # 클래스 이름이 'info_date'인 <span> 태그 아래 텍스트 가져오기
      year_str = soup.find('span', attrs = {'class': 'info_date'}).get_text()
      # 정규 표현식으로 찾은 값 중에 첫 번째 것만 사용
      year = re.findall(r'\d{4}', year_str)[0]
  except AttributeError:
    pass

  return title, author, pub, year
```

```python
def data_fixing(ns_book4):
  """
  잘못된 값을 수정하거나 NaN을 채우는 함수
  :param ns_book4: data_cleaning() 함수에서 전처리된 데이터프레임
  """
  # 도서권수와 대출건수를 int64로 바꾸기
  ns_book4 = ns_book4.astype({'도서권수': 'int64', '대출건수': 'int64'})
  # NaN인 세트 ISBN을 빈 문자열로 바꾸기
  set_isbn_na_rows = ns_book4['세트 ISBN'].isna()
  ns_book4.loc[set_isbn_na_rows, '세트 ISBN'] = ''
  # 발행년도 열에서 연도 네 자리를 추출하여 대체. 나머지 발행년도는 -1로 바꾸기
  ns_book5 = ns_book4.replace({'발행년도': r'.*(\d{4}).*'}, r'\1', regex=True)
  unknown_year = ns_book5['발행년도'].str.contains(r'\D', na=True)
  ns_book5.loc[unknown_year, '발행년도'] = '-1'
  # 발행년도를 int32로 바꾸기
  ns_book5 = ns_book5.astype({'발행년도': 'int32'})
  # 4000년 이상인 경우 2333년을 빼기
  dangun_yy_rows = ns_book5['발행년도'].gt(4000)
  ns_book5.loc[dangun_yy_rows, '발행년도'] = ns_book5.loc[dangun_yy_rows, '발행년도'] - 2333
  # 여전히 4000년 이상인 경우 -1로 바꾸기
  dangun_year = ns_book5['발행년도'].gt(4000)
  ns_book5.loc[dangun_year, '발행년도'] = -1
  # 0~1900년 사이의 발행년도는 -1로 바꾸기
  old_books = ns_book5['발행년도'].gt(0) & ns_book5['발행년도'].lt(1900)
  ns_book5.loc[old_books, '발행년도'] = -1
  # 도서명, 저자, 출판사가 NaN이거나 발행년도가 -1인 행 찾기
  na_rows = ns_book5['도서명'].isna() | ns_book5['저자'].isna() \
  | ns_book5['출판사'].isna() | ns_book5['발행년도'].eq(-1)
  # Yes24 도서 상세 페이지에서 누락된 정보 채우기
  updated_sample = ns_book5[na_rows].apply(get_book_info,
                                           axis=1, result_type = 'expand')
  updated_sample.columns = ['도서명', '저자', '출판사', '발행년도']
  ns_book5.update(updated_sample)
  # 도서명, 저자, 출판사가 NaN이거나 발행년도가 -1인 행 삭제하기
  ns_book6 = ns_book5.dropna(subset=['도서명', '저자', '출판사'])
  ns_book6 = ns_book6[ns_book6['발행년도'] != -1]

  return ns_book6
```
